```python
import sys


import pandas as pd
from os.path import join
from custom_helpers_py.pandas_helpers import read_csv_as_df, save_df_as_json, concat_folder_into_df
from custom_helpers_py.url_formatters import clean_url, get_domain_from_url
from com_filters.helpers import is_url_valid
```

```python
MASS_SCRAPE_AIFT_FOLDER_PATH = "../z_tmp/mass_scrape/aift"
MASS_SCRAPE_AIFT_FP = join(MASS_SCRAPE_AIFT_FOLDER_PATH, "aift-scrape-lists-2024-04-05T21_22_53.071Z-data.csv")

aift_df = read_csv_as_df(MASS_SCRAPE_AIFT_FP)
aift_df = aift_df[["product_url"]]

def is_valid_helper(in_url):
        is_valid_res = is_url_valid(in_url, clean_in_url=True)
        if is_valid_res[0]:
            return False
        return True

aift_df["rejected"] = aift_df["product_url"].apply(is_valid_helper)
aift_df["domain"] = aift_df["product_url"].apply(get_domain_from_url)

rejected_mask = aift_df["rejected"]
aift_df = aift_df[rejected_mask]
aift_rejected_domain_count_df = aift_df["domain"].value_counts()
aift_rejected_domain_count_df = aift_rejected_domain_count_df.reset_index()

save_path = join(MASS_SCRAPE_AIFT_FOLDER_PATH, "domains_rejected_list.json")
save_df_as_json(aift_rejected_domain_count_df, save_path)

gumroad_mask = aift_df["product_url"].str.contains("gumroad.")
aift_gumroad_df = aift_df[gumroad_mask]
num_gumroad_rejected = len(aift_gumroad_df.index)
print("Num gumroad rejected", num_gumroad_rejected)
save_path = join(MASS_SCRAPE_AIFT_FOLDER_PATH, "gumroad_rejected_list.json")
save_df_as_json(aift_gumroad_df, save_path)

aift_df
```
